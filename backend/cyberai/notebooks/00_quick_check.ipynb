```python
import pandas as pd

#Basic Dataset
basic = pd.read_csv("../data/raw/basic.csv")
print("BASIC DATASET")
print("Shape:", basic.shape)
display(basic.head(10))   
print("\nColumns and types:")
display(pd.DataFrame({"column": basic.columns, "dtype": basic.dtypes.astype(str)}))
if "label" in basic.columns:
    print("\nLabel distribution:")
    display(basic["label"].value_counts(dropna=False))


cic = pd.read_csv("../data/raw/cicids2017_cleaned.csv")
print("\n\nCICIDS2017 CLEANED DATASET")
print("Shape:", cic.shape)
display(cic.head(10))   
print("\nColumns and types:")
display(pd.DataFrame({"column": cic.columns, "dtype": cic.dtypes.astype(str)}))
label_col = "Attack Type" if "Attack Type" in cic.columns else "Label"
if label_col in cic.columns:
    print("\nAttack label distribution (top 20):")
    display(cic[label_col].value_counts(dropna=False).head(20))

```
